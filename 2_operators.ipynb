```python
# Break the total amount into denominations. - Input: - Amount = 3700 - Output: - 1000s: 3 - 500s: 1 - Remaining: 200

amount = 3700
a = amount // 1000 #3 
b = amount - (a*1000) # 700
c = b // 500 #1
d = b - (c*500) #200
print(a)
print(c)
print(d)
```

```python
# Convert total seconds into hours, minutes, and seconds. - Input: - Total seconds = 3672 - Output: - Hours: 1 - Minutes: 1 - Seconds: 12

total_seconds = 3672
hours = total_seconds // 3600 #1
remaining_seconds = total_seconds - (hours * 3600) #72
minutes = remaining_seconds // 60 #1
seconds = remaining_seconds - (minutes * 60) #12
print(hours)
print(minutes)
print(seconds)
```

```python
# Calculate the sum of marks in 3 subjects. - Input: - Maths = 85 - Physics = 90 - Chemistry = 88 - Output: - Total marks: 263

maths = 85
physics = 90
chemistry = 88
total_marks = maths + physics + chemistry
print(total_marks)
```
