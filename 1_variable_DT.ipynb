Data type : it used to check what type of data 


Types :

1. Primitive Datatype     -  it store single or multiple data which we cant modify


- (int, float , string , Boolean , complex)

2.Non Primitive Datatype  -  which stores multiple type of data which we can modify

- (List , tuple , set , disctionary)

```python
#Primitive data types

#Integer - A whole number, positive or negative, without decimals. Example: 10, -5, 0

a = 10
b = -20
print(a)
print(b)
print(type(a))
print(type(b))
```

```python
#float - A number that has a decimal point. Example: 3.14, -0.001, 2.0

a = 3.14
b = -0.001
print(a)
print(b)
print(type(a))
print(type(b))
```

```python
#string - A sequence of characters enclosed in quotes. 
# Example: "Hello", 'Python', "123"

a = "Hello"
b = 'Python'
print(a)
print(b)
print(type(a))
print(type(b))
```

```python
#Boolean - A data type that can only have two values: True or False.
# Example: True, False

a = True
b = False
print(a)
print(b)
print(type(a))
print(type(b))
```

```python
#complex numbers

a = 5+6j
print(a)
print(a.real)
print(a.imag)
print(type(a))
```

List : it is a mutable datatype , represented with [](square brackets)


it is ordered , allows Hetrogeneous data , allow duplictaes

```python
list = [1, 2, 3, 4, 5 , "abc" , 1.2]
print(list)
list[0] = "python"
print(list)
print(list[6])
print(type(list))
```

```python
a = (2)
b = ()
c = (20, )
d = (1, 2, 3, 4, 5, "abc", 1.2)
print(a)
print(b)
print(c)
print(d)
print(type(a))
print(type(b))
print(type(c))
print(type(d))
```
